```python
import json

try:
    with open('frontend/public/data/roland.json') as f:
        data = json.load(f)
        products = data.get('products', [])
        print(f"Found {len(products)} products")
        prices = []
        for p in products:
            price = 0
            if 'halilit_data' in p and p['halilit_data'] and 'price' in p['halilit_data']:
                price = p['halilit_data']['price']
            elif 'pricing' in p and p['pricing'] and 'regular_price' in p['pricing']:
                price = p['pricing']['regular_price']
            prices.append(price)
        
        print(f"Prices found: {len([p for p in prices if p > 0])}")
        print(f"Prices sample: {prices[:10]}")
except Exception as e:
    print(f"Error: {e}")
```
